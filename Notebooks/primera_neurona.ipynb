```python
import numpy as np
import matplotlib.pyplot as plt
```

Una neurona necessita decidir si "s'activa" o no.
La funció sigmoid converteix qualsevol número en un valor entre 0 i 1.

**Equació:**

$$\sigma(x) = \frac{1}{1 + e^{-x}}$$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Visualització de la funció sigmoid
x = np.linspace(-10, 10, 100)
plt.plot(x, sigmoid(x), color='steelblue')
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.title('Funció d\'activació sigmoid')
plt.xlabel('Entrada (x)')
plt.ylabel('Sortida σ(x)')
plt.grid(True, alpha=0.3)
plt.show()
```

Una neurona fa tres coses:

1. Multiplica cada entrada pel seu pes: $x \cdot w$
2. Suma un biaix: $+ b$
3. Aplica la funció d'activació: $\sigma(x \cdot w + b)$

**Equació completa:**

$$\text{sortida} = \sigma\left(\sum_{i} x_i \cdot w_i + b\right)$$

```python
def neurona(entrades, pesos, biaix):
    z = np.dot(entrades, pesos) + biaix  # combinació lineal
    print(f"Combinació lineal (z): {z:.4f}")
    return sigmoid(z)                     # activació

# Exemple: una neurona amb 3 entrades
entrades = np.array([0.2, 0.8, 0.8])
pesos    = np.array([0.4, -0.5, 0.1])   # iniciats manualment
biaix    = 0.1

sortida = neurona(entrades, pesos, biaix)
print(f"Sortida de la neurona: {sortida:.4f}")
print(f"La neurona {'S\'ACTIVA' if sortida > 0.5 else 'NO s\'activa'} (frontera 0.5)")
```
